```python
import pandas as pd
import matplotlib.pyplot as plt
election = pd.read_csv('election_results.csv')
election
```

```python
tableau20 = [(31, 119, 180), (44, 160, 44), (148, 103, 189), (227, 119, 194), (188, 189, 34), ] 
for i in range(len(tableau20)):    
    r, g, b = tableau20[i]    
    tableau20[i] = (r / 255., g / 255., b / 255.) 
plt.figure(figsize=(12, 9)) 
ax = plt.subplot(111)
ax.spines["top"].set_visible(False)    
ax.spines["bottom"].set_visible(False)    
ax.spines["right"].set_visible(False)    
ax.spines["left"].set_visible(False) 
ax.get_xaxis().tick_bottom()    
ax.get_yaxis().tick_left()
plt.ylim(0, 50)    
plt.xlim(1997, 2017)
plt.yticks(range(0, 51, 10), [str(x) + "%" for x in range(0, 51, 10)], fontsize=14)    
plt.xticks(fontsize=14)
for y in range(10, 51, 10):    
    plt.plot(range(1997, 2017), [y] * len(range(1997, 2017)), "--", lw=0.5, color="black", alpha=0.3)
plt.tick_params(axis="both", which="both", bottom="off", top="off",    
            labelbottom="on", left="off", right="off", labelleft="on")
countries = ['Austria', 'France', 'Germany', 'Netherlands', 'United Kingdom']
for rank, column in enumerate(countries):    
    plt.plot(election.Year.values,    
            election[column.replace("\n", " ")].values,    
            lw=2.5, color=tableau20[rank])  
    y_pos = election[column.replace("\n", " ")].values[-1] - 0.5
    if column == "Austria":    
        y_pos += 0.1    
    elif column == "France":    
        y_pos += 0.1    
    elif column == "Germany":    
        y_pos += 0.1    
    elif column == "Netherlands":    
        y_pos += 1    
    elif column == "United Kingdom":    
        y_pos -= 1 
    plt.text(2017.5, y_pos, column, fontsize=14, color=tableau20[rank]) 

plt.show()
plt.savefig('election_results_graph.png', dpi=300)
```
