```python
import pandas as pd
import matplotlib.pyplot as plt
hatecrime = pd.read_csv('hate_crime_numbers.csv')
hatecrime
```

```python
auscolor = [(31, 119, 180)] 
for i in range(len(auscolor)):    
    r, g, b = auscolor[i]    
    auscolor[i] = (r / 255., g / 255., b / 255.) 
plt.figure(figsize=(12, 9)) 
ax = plt.subplot(111)
ax.spines["top"].set_visible(False)    
ax.spines["bottom"].set_visible(False)    
ax.spines["right"].set_visible(False)    
ax.spines["left"].set_visible(False) 
ax.get_xaxis().tick_bottom()    
ax.get_yaxis().tick_left()
plt.ylim(0, 500)    
plt.xlim(2009, 2015)
plt.yticks(range(0, 501, 50), [str(x) for x in range(0, 501, 50)], fontsize=14)    
plt.xticks(fontsize=14)
for y in range(0, 501, 50):    
    plt.plot(range(2009, 2016), [y] * len(range(2009, 2016)), "--", lw=0.5, color="black", alpha=0.3)
plt.tick_params(axis="both", which="both", bottom="off", top="off",    
            labelbottom="on", left="off", right="off", labelleft="on")
countries = ['Austria']
for rank, column in enumerate(countries):    
    plt.plot(hatecrime.Year.values,    
            hatecrime[column.replace("\n", " ")].values,    
            lw=2.5, color=auscolor[rank]) 
plt.show()
```
