```python
with open('Data.csv') as f:
    myfile = f.read()
import pandas as pd
df = pd.read_csv('Data.csv')
df.head()
```

```python
from bokeh.plotting import *

source=ColumnDataSource(
    data=dict(
lats = list(df['lat']),
longs = list(df['neg_long']), 
bandb=list(df['black/brown percentage']),
white=list(df['white percentage']),
    )
)

lats = list(df['lat'])
longs = list(df['neg_long']) 
bandb=list(df['black/brown percentage'])
white=list(df['white percentage'])
      
from bokeh.sampledata import us_states
from bokeh.models import HoverTool

us_states = us_states.data.copy()

#del laska and hawaii
del us_states["HI"]
del us_states["AK"]

# separate latitude and longitude points for the borders
#   of the states.
state_xs = [us_states[code]["lons"] for code in us_states]
state_ys = [us_states[code]["lats"] for code in us_states]

hover = HoverTool(
        tooltips=[
            ("lats", "@lats"),
            ("longs", "@longs"),
            ("bandb", "@bandb"),
        ]
    )

# init figure
p = figure(title="US Black and Brown Population Map: Move Mouse over area to view more!",
           toolbar_location="left", tools=[hover,"pan","save","box_zoom"], plot_width=1100, plot_height=700)

# Draw state lines
p.patches(state_xs, state_ys, fill_alpha=0.0,
    line_color="#884444", line_width=1.5)

#Plot Color Bubbles for black and brown people
p.circle(longs, lats, size=.5, color='blue', alpha=bandb, source=source)

#Output 
output_file("US Black and Brown Population Map.html")

#show results
show(p)
```
